```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression #To train ML model that can predict salaries based on input features like job title, skills, etc
from sklearn.preprocessing import OneHotEncoder #Imports a tool that turns text categories into numbers (0s and 1s).
from sklearn.preprocessing import StandardScaler #. It transforms data so that the mean becomes 0 and the standard deviation becomes 1.
from sklearn.feature_extraction.text import CountVectorizer #Imports CountVectorizer which turns text (skills) into numerical values
from sklearn.compose import ColumnTransformer #Allows us to apply different preprocessing methods to different columns.
from sklearn.pipeline import Pipeline #It connect everything (preprocessing + model) in one clean pipeline
from sklearn.metrics import mean_squared_error #Brings in a way to measure how good your model is.
```

```python
df=df[['job_title','company_location','experience_level','required_skills','years_experience','education_required','salary_usd']]
```

```python
print(type(df))
```

```python
print(df)
x=df[['job_title','company_location','experience_level','required_skills','years_experience']]
y=df['salary_usd']
```

```python
categorical_features=['job_title','company_location','experience_level']
text_features='required_skills'
```

```python
#"Use different transformations for different types of data (columns)."
#'cat': just a name for this step
#OneHotEncoder(...): changes text (like "US", "India", "UK") into numbers (0s and 1s)
#categorical_features: the columns you want to apply this to — example: job_title, company_location, etc.
preprocessor=ColumnTransformer(transformers=[('cat',OneHotEncoder(handle_unknown='ignore'),categorical_features),
                                            ('text',CountVectorizer(tokenizer=lambda x:x.split(', ')),'required_skills'),#CountVectorizer(...): converts a list of words (skills) into numbers (like word counts)
                                             ('num', StandardScaler(), ['years_experience'])])#tokenizer=lambda x: x.split(', '): tells it how to split skills — by comma and space, like:
```

```python
#way to connect multiple steps of data processing and modeling together in one object
model=Pipeline(steps=[('preprocessor',preprocessor),('regressor',LinearRegression())])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) # 80% training and 20% testing data
```

```python
model.fit(x_train,y_train) #This trains the machine learning model using the training data.
```

```python
y_pred=model.predict(x_test) #Predicts salaries for test data.
```

```python
rmse=mean_squared_error(y_test,y_pred) #calculates model accuracy by comparing actual and predicted salaries
```

```python
print(round(rmse,2))
```
